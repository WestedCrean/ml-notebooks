```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
plt.rcParams["figure.figsize"] = (14,12)
df = pd.read_excel('01-20.xls')
df = df.set_index(df['Unnamed: 0'])
display(df)
print(df.columns)
```

```python
# 1. Czy częstotliwość przeskoków zależy od napięcia elektrofiltru?
#
# Owszem, przy napięciu około 35 V oraz powyżej 42 V aż do 50 V częstość przeskoków rośnie dimetralnie

df1 = df.sort_values(by=['Napięcie elektrofiltru - pole 1','Napięcie elektrofiltru - pole 2','Napięcie elektrofiltru - pole 3','Napięcie elektrofiltru - pole 4',])

df['Napięcie elektrofiltru'] = df.iloc[:,14:18].mean(axis=1)
df['Częstość przeskoków'] = df.iloc[:,19:22].mean(axis=1)
df['Prąd elektrofiltru'] = df.iloc[:, 10:13].mean(axis=1)

slice_df = df[['Napięcie elektrofiltru','Częstość przeskoków']].dropna().sort_values(by='Napięcie elektrofiltru')


slice_df.plot(x='Napięcie elektrofiltru', y='Częstość przeskoków')
```
